## This file shows how to make use of the data_path.csv file to access the  wav files

```
import pandas as pd
import numpy as np
import os

import librosa
import librosa.display
import seaborn as sns
import matplotlib.pyplot as plt
```

```
data_path = pd.read_csv('data_path.csv')
data_path.head()
```

```

print(data_path.Emotions.value_counts())
```

```
# Set the type of the Emotions column to category
# Plot out the count of each emotion in all the files in the dataset

plt.title('Count of Emotions', size=16)
ax = sns.countplot(data_path, x='Emotions')
plt.ylabel('Count', size=12)
plt.xlabel('Emotions', size=12)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

```
# Sample wav files and their corresponding emotions
print(data_path['Path'][0], data_path['Emotions'][0])
print(data_path['Path'][98], data_path['Emotions'][98])
print(data_path['Path'][482], data_path['Emotions'][482])
print(data_path['Path'][868], data_path['Emotions'][868])
print(data_path['Path'][290], data_path['Emotions'][290])
print(data_path['Path'][1250], data_path['Emotions'][1250])
print(data_path['Path'][714], data_path['Emotions'][714])
print(data_path['Path'][1058], data_path['Emotions'][1058])
```

```
data_neutral, sr_neutral = librosa.load(os.getcwd() + '\\wav_file\\'+data_path['Path'][0])
data_calm, sr_calm = librosa.load(os.getcwd() + '\\wav_file\\'+data_path['Path'][98])
data_sad, sr_sad = librosa.load(os.getcwd() + '\\wav_file\\'+data_path['Path'][482])
data_fear, sr_fear = librosa.load(os.getcwd() + '\\wav_file\\'+data_path['Path'][868])
data_happy, sr_happy = librosa.load(os.getcwd() + '\\wav_file\\'+data_path['Path'][290])
data_surprise, sr_surprise = librosa.load(os.getcwd() + '\\wav_file\\'+data_path['Path'][1250])
data_angry, sr_angry = librosa.load(os.getcwd() + '\\wav_file\\'+data_path['Path'][1448])
data_disgust, sr_disgust = librosa.load(os.getcwd() + '\\wav_file\\'+data_path['Path'][1058])
```

```
#Wave plot of the different emotions
plt.figure(figsize=(8,2))
plt.title('Wave for calm', size=20)
librosa.display.waveshow(data_calm, sr=sr_calm)
plt.figure(figsize=(8,2))
plt.title('Wave for sad', size=20)
librosa.display.waveshow(data_sad, sr=sr_sad)
plt.figure(figsize=(8,2))
plt.title('Wave for Neutral', size=20)
librosa.display.waveshow(data_neutral, sr=sr_neutral)
plt.figure(figsize=(8,2))
plt.title('Wave for Fear', size=20)
librosa.display.waveshow(data_fear, sr=sr_fear)
plt.figure(figsize=(8,2))
plt.title('Wave for Happy', size=20)
librosa.display.waveshow(data_happy, sr=sr_happy)
plt.figure(figsize=(8,2))
plt.title('Wave for Surprise', size=20)
librosa.display.waveshow(data_surprise, sr=sr_surprise)
plt.figure(figsize=(8,2))
plt.title('Wave for Angry', size=20)
librosa.display.waveshow(data_angry, sr=sr_angry)
plt.figure(figsize=(8,2))
plt.title('Wave for Disgust', size=20)
librosa.display.waveshow(data_disgust, sr=sr_disgust)
```

```
# Spectrogram of the different emotions
plt.figure(figsize=(8,2))
plt.title('Spectrogram for calm', size=20)
x = librosa.stft(data_calm)
Xdb = librosa.amplitude_to_db(abs(x))
librosa.display.specshow(Xdb, sr=sr_calm, x_axis='time', y_axis='hz')
plt.figure(figsize=(8,2))
plt.title('Spectrogram for sad', size=20)
x = librosa.stft(data_sad)
Xdb = librosa.amplitude_to_db(abs(x))
librosa.display.specshow(Xdb, sr=sr_sad, x_axis='time', y_axis='hz')
plt.figure(figsize=(8,2))
plt.title('Spectrogram for Neutral', size=20)
x = librosa.stft(data_neutral)
Xdb = librosa.amplitude_to_db(abs(x))
librosa.display.specshow(Xdb, sr=sr_neutral, x_axis='time', y_axis='hz')
plt.figure(figsize=(8,2))
plt.title('Spectrogram for Fear', size=20)
x = librosa.stft(data_fear)
Xdb = librosa.amplitude_to_db(abs(x))
librosa.display.specshow(Xdb, sr=sr_fear, x_axis='time', y_axis='hz')
plt.figure(figsize=(8,2))
plt.title('Spectrogram for Happy', size=20)
x = librosa.stft(data_happy)
Xdb = librosa.amplitude_to_db(abs(x))
librosa.display.specshow(Xdb, sr=sr_happy, x_axis='time', y_axis='hz')
plt.figure(figsize=(8,2))
plt.title('Spectrogram for Surprise', size=20)
x = librosa.stft(data_surprise)
Xdb = librosa.amplitude_to_db(abs(x))
librosa.display.specshow(Xdb, sr=sr_surprise, x_axis='time', y_axis='hz')
plt.figure(figsize=(8,2))
plt.title('Spectrogram for Angry', size=20)
x = librosa.stft(data_angry)
Xdb = librosa.amplitude_to_db(abs(x))
librosa.display.specshow(Xdb, sr=sr_angry, x_axis='time', y_axis='hz')
plt.figure(figsize=(8,2))
plt.title('Spectrogram for Disgust', size=20)
x = librosa.stft(data_disgust)
Xdb = librosa.amplitude_to_db(abs(x))
librosa.display.specshow(Xdb, sr=sr_disgust, x_axis='time', y_axis='hz')
```

## Pre-emphasis
### Pre-emphasis is a way to amplify the high frequencies and reduce the lower frequencies in the audio signal.

```
# using the calm wav file as an example

data_calm_filter = librosa.effects.preemphasis(data_calm)
# original signal
plt.figure(figsize=(8,2))
plt.title('Original Signal (Calm)', size=20)
S_orig = librosa.amplitude_to_db(np.abs(librosa.stft(data_calm)), ref=np.max)
librosa.display.specshow(S_orig, y_axis='hz', x_axis='time')
# filtered signal
plt.figure(figsize=(8,2))
plt.title('Filtered Signal (Calm)', size=20)
S_preemph = librosa.amplitude_to_db(np.abs(librosa.stft(data_calm_filter)), ref=np.max)
librosa.display.specshow(S_preemph, y_axis='hz', x_axis='time')
```

## Understanding female vs male wav files using MFCC

```
# Male happy
data_mhappy, sr_mhappy = librosa.load(os.getcwd() + '\\wav_file\\'+data_path['Path'][290])
male = librosa.feature.mfcc(y=data_mhappy, sr=sr_mhappy, n_mfcc=13)
male = np.mean(librosa.feature.mfcc(y=data_mhappy, sr=sr_mhappy, n_mfcc=13), axis=0)
# Female happy
data_fcalm, sr_fcalm = librosa.load(os.getcwd() + '\\wav_file\\'+data_path['Path'][291])
female = librosa.feature.mfcc(y=data_fcalm, sr=sr_fcalm, n_mfcc=13)
female = np.mean(librosa.feature.mfcc(y=data_fcalm, sr=sr_fcalm, n_mfcc=13), axis=0)

plt.figure(figsize=(15, 15))
plt.subplot(3,1,1)
plt.plot(female, label='female')
plt.plot(male, label='male')
plt.legend()
```

```
# Male angry
data_mangry, sr_mangry = librosa.load(os.getcwd() + '\\wav_file\\'+data_path['Path'][694])
male = librosa.feature.mfcc(y=data_mangry, sr=sr_mangry, n_mfcc=13)
male = np.mean(librosa.feature.mfcc(y=data_mangry, sr=sr_mangry, n_mfcc=13), axis=0)
# Female angry
data_fcalm, sr_fcalm = librosa.load(os.getcwd() + '\\wav_file\\'+data_path['Path'][695])
female = librosa.feature.mfcc(y=data_fcalm, sr=sr_fcalm, n_mfcc=13)
female = np.mean(librosa.feature.mfcc(y=data_fcalm, sr=sr_fcalm, n_mfcc=13), axis=0)

plt.figure(figsize=(15, 15))
plt.subplot(3,1,1)
plt.plot(female, label='female')
plt.plot(male, label='male')
plt.legend()
```
